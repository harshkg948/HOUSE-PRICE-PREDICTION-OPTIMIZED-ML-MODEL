```python
df = df.dropna(subset=['SalePrice'])
```

```python
std_lot_area = StandardScaler()
std_year_built = StandardScaler()
std_year_remod = StandardScaler()
```

```python
df['LotArea'] = std_lot_area.fit_transform(df[['LotArea']])
df['YearBuilt'] = std_year_built.fit_transform(df[['YearBuilt']])
df['YearRemodAdd'] = std_year_remod.fit_transform(df[['YearRemodAdd']])
```

```python
std_ttl_bsmt_sf = StandardScaler()
df['TotalBsmtSF'] = std_ttl_bsmt_sf.fit_transform(df[['TotalBsmtSF']])
```

```python
from sklearn.preprocessing import LabelEncoder
mszoning_enco = LabelEncoder()
bldg_type_enco = LabelEncoder()
lot_config_enco = LabelEncoder()
df['MSZoning_enco'] = mszoning_enco.fit_transform(df['MSZoning'])
df['BldgType_enco'] = bldg_type_enco.fit_transform(df['BldgType'])
df['LotConfig_enco'] = lot_config_enco.fit_transform(df['LotConfig'])
```

```python

df[['LotConfig','LotConfig_enco']].drop_duplicates()
```

```python
df[['MSZoning','MSZoning_enco']].drop_duplicates()
```

```python
df[['BldgType','BldgType_enco']].drop_duplicates()
```

```python
df.info()
```

```python
features = [
    'MSSubClass',
    'MSZoning_enco',
    'LotArea',
    'LotConfig_enco',
    'BldgType_enco',
    'OverallCond',
    'YearBuilt',
    'YearRemodAdd',
    'TotalBsmtSF'   
]
X = df[features]
```

```python
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,train_size=0.7)
```

```python
df
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=200,random_state=42)
model.fit(X_train,y_train)
```

```python
y_true = df['SalePrice']
```

```python
answer = model.predict(X_test)
```

```python
answer
```

```python
y_pred = model.predict(X_train)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_train,y_pred)
```
